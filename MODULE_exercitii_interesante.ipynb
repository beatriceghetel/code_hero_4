## Jocul ❌ si ⭕ (matrici si foarte putina matematica)

Cand jucati jocul X si 0 cand erati mici, foloseati fara sa stiti un concept matematic: matricea. Matricea e ca un tabel de fapt. La un tabel avem linii si coloane. Evident, liniile si coloanele sunt numerotate de la 1 ... n dar in programare le vom numerota de la 0 ... n-1. Astfel, elementul din colt stanga sus va fi pe linia 0 coloana 0. Colt dreapta sus va fi linia 0 coloana n-1 (0 si 2 pentru x si 0). Colt stanga jos linia n-1 coloana 0.

Aplicatiile stocheaza in memorie informatii legate de utilizatori si retele de socializare folosind structuri de date speciale precum grafuri. Un graf este un set de puncte unite intre ele prin linii.

Daca avem 4 persoane pe Facebook, Ion, Ana, Diana si Vasile, ei vor fi reprezentati ca **puncte**.

Daca Ion si Ana sunt prieteni pe Facecbook, intre ei va fi trasa o **linie** pentru a arata ca ei sunt prieteni si se cunosc. Daca Diana ii cunoaste si pe Ion si pe Ana, va fi trasa o linie intre Ion - Diana si Ana - Diana.

Daca Vasile nu are prieteni, nu va avea nici o linie care-l uneste de alti utilizatori (cand iti creezi cont nou, pana adaugi prieteni de exemplu)

```python
import networkx

retea_sociala = networkx.Graph() # cream o retea sociala noua, goala

# Adaugam "noduri" - adica persoane in cazul nostru
retea_sociala.add_nodes_from(['Ion','Ana','Diana','Vasile', 'Dan', 'Vlad', 'Andreea'])

# Adaugam linii - unim punctele intre ele
# Principiul intr-o retea de socializare este sa unim acei utilizatori care se cunosc
# si se afla intro relatie de prietenie
# Dana Ion si Ana sunt prieteni pe Facebook, reprezentam asta trasand o linie de la Ion la Ana
retea_sociala.add_edges_from ([('Ion','Ana'), ('Diana','Ion'), ('Ana', 'Diana'), ('Vasile', 'Ion'), ('Andreea', 'Diana'),('Dan', 'Vlad'), ('Vlad', 'Ion')])

networkx.draw(retea_sociala, with_labels = True)
```
